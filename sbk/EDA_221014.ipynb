## 문제인 정권 주간 아파트 매매 지수

------

```python
moon = pd.read_excel('./data/주간_매매가격지수_종합.xlsx')
```

```python
moon
```

```python
moon = moon[moon.columns.drop(list(moon.filter(regex='Unnamed')))]
moon
```

```python
moon_col = moon.columns[1:]
moon_col
```

```python
moon_val = moon.values[:, 1:]
moon_val
```

```python
moon = pd.DataFrame(index=moon_col, data=moon_val.T)
moon
```

```python
def change_date_to_time2(x):
    x = x.replace("년 ", "-")
    x = x.replace("월" , "-")
    x = x.replace("일", " ")
    return x
```

```python
moon.reset_index()["index"].apply(change_date_to_time2)
```

```python
moon = moon.reset_index()
moon["index"] = moon["index"].apply(change_date_to_time2)
moon
```

```python
moon.columns = ['Date', 'Country', 'Seoul', 'Seocho', 'Gangnam', 'Songpa']
moon
```

```python
import plotly.express as px
df = moon
fig = px.line(df, x="Date", y=df.columns,
              hover_data={"Date": "%Y%m"},
              title='moon_RealEstate')
fig.update_xaxes(
    dtick="M2",
    tickformat="%Y%m",
    ticklabelmode="period")
fig.show()
```

## 주요국 경기 선행지수

------

```python
economic_idx = pd.read_csv('./data/경기선행지수.csv', encoding='cp949')
economic_idx
```

```python
economic_idx_col = economic_idx.columns[1:]
economic_idx_col
```

```python
economic_idx_val = economic_idx.values[:, 1:]
economic_idx_val
```

```python
economic_idx = pd.DataFrame(index=economic_idx_col, data=economic_idx_val.T)
economic_idx
```

```python
economic_idx.index = pd.to_datetime(economic_idx.index, format = '%Y/%m')
economic_idx.index
```

```python
economic_idx = economic_idx.reset_index()
economic_idx["index"] = economic_idx["index"]
economic_idx
```

```python
economic_idx.columns = ['Date', 'Japan', 'Korea', 'UK', 'USA', 'Euro', 'China']
economic_idx['Date'].info
```

```python
plt.figure(figsize = (30,15))
# sns.set(font_scale=2)

sns.lineplot(data=economic_idx, x='Date', y='Japan')
sns.lineplot(data=economic_idx, x='Date', y='Korea')
sns.lineplot(data=economic_idx, x='Date', y='UK')
sns.lineplot(data=economic_idx, x='Date', y='USA')
sns.lineplot(data=economic_idx, x='Date', y='Euro')
sns.lineplot(data=economic_idx, x='Date', y='China')

plt.legend(labels = ['Japan', 'Korea', 'UK', 'USA', 'Euro', 'China'], fontsize=20)
plt.xlabel('Date', fontsize=20)
plt.ylabel('economic', fontsize=20)
plt.show
```

## 월간 종합 주택 매매지수
------------

```python
country
```

```python
country = country[country.columns.drop(list(country.filter(regex='Unnamed')))]
country
```

```python
country_col = country.columns
country_col
```

```python
country_val = country.values
country_val
```

```python
# country=country.transpose()
# country
country = pd.DataFrame(index=country_col, data=country_val.T)
country
```

```python
def change_date_to_time(x):
    x = x.replace("년 ", "-")
    x = x.replace("월", " ")
#     x = x.replace("일", " ")
    return x
```

```python
country.reset_index()["index"].apply(change_date_to_time)
```

```python
country = country.reset_index()
country["index"] = country["index"].apply(change_date_to_time)
country
```

```python
country.columns = ['Date', 'Price_index']
country
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(data=country, x='Date', y='Price_index')
plt.show()
```

## 주요국 경기 선행지수

------

```python
economic_idx = pd.read_csv('./data/경기선행지수.csv', encoding='cp949')
economic_idx
```

```python
economic_idx_col = economic_idx.columns[1:]
economic_idx_col
```

```python
economic_idx_val = economic_idx.values[:, 1:]
economic_idx_val
```

```python
economic_idx = pd.DataFrame(index=economic_idx_col, data=economic_idx_val.T)
economic_idx
```

```python
economic_idx.index = pd.to_datetime(economic_idx.index, format = '%Y/%m')
economic_idx.index
```

```python
economic_idx = economic_idx.reset_index()
economic_idx["index"] = economic_idx["index"]
economic_idx
```

```python
economic_idx.columns = ['Date', 'Japan', 'Korea', 'UK', 'USA', 'Euro', 'China']
economic_idx['Date'].info
```

```python
plt.figure(figsize = (30,15))
# sns.set(font_scale=2)

sns.lineplot(data=economic_idx, x='Date', y='Japan')
sns.lineplot(data=economic_idx, x='Date', y='Korea')
sns.lineplot(data=economic_idx, x='Date', y='UK')
sns.lineplot(data=economic_idx, x='Date', y='USA')
sns.lineplot(data=economic_idx, x='Date', y='Euro')
sns.lineplot(data=economic_idx, x='Date', y='China')

plt.legend(labels = ['Japan', 'Korea', 'UK', 'USA', 'Euro', 'China'], fontsize=20)
plt.xlabel('Date', fontsize=20)
plt.ylabel('economic', fontsize=20)
plt.show
```

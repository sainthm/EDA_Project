```python
# import module
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# import plotly
import plotly.express as px
```

```python
# data check - aspoa
aspoa = pd.read_csv("./EDA_file/aspoa.csv")
```

```python
aspoa.head(15)
```

```python
# DataFrame
df = pd.DataFrame(aspoa)
df
```

```python
df.shape
```

```python
df_all = df.transpose()
df_all
```

```python
# 전국 > df_nation
df_nation = df_all.iloc[:,[0, 1, 2, 3, 4]]
df_nation
```

```python
# 전국 데이터 전처리
df_nation.columns = ["nano", "micro", "small", "medium", "large"]
```

```python
# 전국 데이터 시각화
plt.figure(figsize=(60, 30))
fig = px.line(df_nation)
fig.show()
```

```python
# 서울 > df_seoul
df_seoul = df_all[[5,6,7,8,9]]
df_seoul
```

```python
# 서울 전처리
df_seoul.columns = ["nano", "micro", "small", "medium", "large"]
```

```python
# 서울 데이터 시각화
plt.figure(figsize=(60, 30))
fig = px.line(df_seoul)
fig.show()
```

```python
# 수도권 > df_capital
df_capital = df_all[[10, 11, 12, 13, 14]]
df_capital
```

```python
# 수도권 전처리
df_capital.columns = ["nano", "micro", "small", "medium", "large"]
```

```python
# 수도권 시각화
plt.figure(figsize=(60, 30))
fig = px.line(df_capital)
fig.show()
```
